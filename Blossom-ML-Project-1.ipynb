# Question 3 

```python
##In which provinces does drug-related crime occur more than 1000 times a year (on average).
Drug_relate=sa[(sa['Category']=='Drug-related crime')]
dw=Drug_relate.groupby('Province').mean()
dw['Average_Total_Crime']=dw['2005-2006']+dw['2006-2007']+dw['2007-2008']+dw['2008-2009']+dw['2009-2010']+dw['2010-2011']+dw['2011-2012']+dw['2012-2013']+dw['2013-2014']+dw['2014-2015']+dw['2015-2016']
dw
Province=dw[(dw['Average_Total_Crime'] > 1000)]
Province[['Average_Total_Crime']]
```

```python
##Which Province has the highest number of stations? 
sa.groupby(["Province"])[['Station']].count()
```

### Eastern Cape has the highest number of stations

```python
##What is the average crime rate per station in Eastern Cape province?
East_Crime=sa[(sa['Province']=='Eastern Cape')]
East_Crime_Prov=East_Crime.groupby(["Station"]).mean()
East_Crime_Prov['Average_Crime_rate']=(East_Crime_Prov['2005-2006']+East_Crime_Prov['2006-2007']+East_Crime_Prov['2007-2008']+East_Crime_Prov['2008-2009']+East_Crime_Prov['2009-2010']+East_Crime_Prov['2010-2011']+East_Crime_Prov['2011-2012']+East_Crime_Prov['2012-2013']+East_Crime_Prov['2013-2014']+East_Crime_Prov['2014-2015']+East_Crime_Prov['2015-2016'])/11
East_Crime_Prov[['Average_Crime_rate']]
```

# Question 5 

```python
#Selecting relevant columns for plot
co_ty=sa[sa.columns[5:17]]
coo=co_ty.loc[(co_ty.Category == 'Drug-related crime') | (co_ty.Category == 'Bank robbery')| (co_ty.Category == 'Murder ') |(co_ty.Category == 'Truck hijacking')]
Co_co_ty=coo.groupby('Category').sum()
#Transposing dataframe
Transpose=Co_co_ty.T
Transpose
Transpose.plot(kind='box')
plt.show()
```

```python
#Boxplots of some crime type distributions across time.
sns.boxplot(x='Year', y='Murder', width=1,data=Transpose)
plt.show()
```

```python
#Kernel density estimates of some crime type distributions across time.
sns.distplot(Transpose['Bank robbery'], hist=True, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})
```

```python
#Histogram of the distribution of some crime type across provinces.
Hist_Crime=sa[(sa['Category']=='Robbery at residential premises')]
Hist_Crime
Hist_Crime_Hist=Hist_Crime.groupby(["Province"])[['Category']].count()
sns.distplot(Hist_Crime_Hist['Category'])
```
